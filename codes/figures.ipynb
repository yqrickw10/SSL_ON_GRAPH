```python
import numpy
import matplotlib.pyplot as plt
```

```python
k = [2,10,20,30,50,70,100,150,200,300,435]
accuracy = [0.8488,0.8721,0.7909,0.5698,0.6674,0.9326,0.9349,0.8977,0.8953,0.9349,0.9372]
```

```python
fig, ax = plt.subplots()
ax.plot(k, accuracy,linestyle='--', marker='o', color='r',markersize=4)

ax.set(xlabel='Number of eigenpairs', ylabel='Classification Accuracy',
       title='Relation between number of eigenpairs and accuracy')
ax.grid()

fig.savefig("Accuaracy-number of eigenpairs relation.png",dpi=240)
plt.show()
```

```python
n = [8,10,12,14,16]
a = [0.8698,0.9256,0.9233,0.9349,0.9372]

fig, ax = plt.subplots()
ax.plot(n, a,linestyle='--', marker='o', color='b',markersize=4)

ax.set(xlabel='Number of features', ylabel='Classification Accuracy',
       title='Relation between number of features and accuracy')
ax.grid()

fig.savefig("Accuaracy-number of features relation.png",dpi=240)
plt.show()
```
